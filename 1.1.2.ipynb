#Calculator with if elif else

```python
print('Press 1 : for add')
print('Press 2 : for subtract')
print('Press 3 : for multiply')
print('Press 4 : for divide')
user_input = int(input('Please Enter Something : '))
print(user_input)
print(type(user_input))
#we need to know Which number and which number will do the operation?
#must be change integer because 1
first_Num = int(input('Please Enter 1st Number : '))
Second_Num = int(input('Please Enter 2nd Number : '))
if user_input == 1:
  result = first_Num + Second_Num
elif user_input == 2:
  result = first_Num - Second_Num
elif user_input == 3:
  result = first_Num * Second_Num
elif user_input == 4:
  result = first_Num / Second_Num
else:
  print('You must enter 1/2/3/4')

print('The number of {} {} {} is {}'. format(first_Num, user_input, Second_Num,  result ))
```

```python
def add(a, b): #functional Header, (a,b = parameter)
  return (a + b) #function body
def sub(a, b):
  return (a - b)
def  multi(a, b):
  return (a * b)
def divi(a, b):
  return (a / b)
fst_num = int(input('Please Enter 1st Number : '))
Snd_num = int(input('Please Enter 2nd Number : '))
print('Adding is :', add(fst_num, Snd_num))
print('Subtracting is :', sub(fst_num, Snd_num))
print('Multiplying is :', multi(fst_num, Snd_num))
print('Dividing is :', divi(fst_num, Snd_num))
#def calculator
```
